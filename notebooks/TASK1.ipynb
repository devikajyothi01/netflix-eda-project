```python
df = df.copy()
```

```python
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Not Available")
df['country'] = df['country'].fillna("Unknown")
df['rating'] = df['rating'].fillna("TV-MA")
```

```python
#MOVIES VS TV SHOWS
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='type',data=df)
plt.title("Movies v/s TV shows in Netflix")
plt.show()
```

Netflix contain more Movies than TV shows

```python
#Content Release Trend
sns.histplot(df['release_year'],bins=20)
plt.title("Distribution of Release Year")
plt.show()
```

Content production increased significantly after 2015

```python
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index)
plt.title("Content Ratings Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,5))

df[df['country'] != "Unknown"]['country'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Content Producing Countries")
plt.xlabel("Country")
plt.ylabel("Count")

plt.xticks(rotation=45)
plt.show()
```

```python
top_genres = df['listed_in'].value_counts().head(10)

top_genres.plot(kind='barh')
plt.title("Top Genres on Netflix")
plt.show()
```

Key Findings:

1. Majority of content on Netflix are Movies.
2. Content production increased after 2015.
3. TV-MA is the most common rating.
4. USA is the leading content producer.
5. Drama is one of the most popular genres.
